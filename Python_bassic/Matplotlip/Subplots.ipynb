# Vẽ nhiều biểu đồ

+ plt.tight_layout(): tự động điều chỉnh khoảng cách các biểu đồ, để tránh việc chồng lên nhau


```python
import matplotlib.pyplot as plt

#biểu đồ 1
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# vẽ biểu đồ phân tán
plt.subplot(2, 1, 1) # 2 hàng, 1 cột, vị trí 1
plt.scatter(x, y, color='red')

plt.title("Biểu đồ phân tán")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()

#biểu đồ 2
plt.subplot(2, 1, 2)  # 2 hàng, 1 cột, vị trí 2
plt.plot([0, 1, 2, 3], [30, 25, 20, 10])
plt.title("Biểu đồ 2")

plt.tight_layout()  # Để các biểu đồ không chồng lấn
plt.show()
```

+ `plt.subplots(2, 2)`: Hàm plt.subplots() tạo ra một "figure" và một mảng các "axes" (các trục biểu đồ con)

+ fig:
     Đây là "figure" chính bao gồm tất cả các biểu đồ con. Có thể coi figure như một tờ giấy hoặc khung chứa để vẽ các biểu đồ con bên trong.
+ axs:
    Đây là một mảng 2 chiều chứa các "axes" - mỗi axes là một biểu đồ con.


```python
import matplotlib.pyplot as plt

# Vẽ một biểu đồ mẫu
fig, axs = plt.subplots(2, 2)  # Tạo 2x2 grid cho các biểu đồ
axs[0, 0].plot([1, 2, 3], [1, 4, 9])
axs[0, 1].plot([1, 2, 3], [1, 2, 3])
axs[1, 0].plot([1, 2, 3], [1, 0, 1])
axs[1, 1].plot([1, 2, 3], [3, 2, 1])

plt.tight_layout()
plt.show()
```
