# Vẽ biểu đồ phân tán 

#### ` plt.scatter()`

```python
import matplotlib.pyplot as plt

# dữ liệu cho biểu đồ phân tán
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# vẽ biểu đồ phân tán
plt.scatter(x, y, color='red')

plt.title("Biểu đồ phân tán")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

+ alpha=0.5: Độ trong suốt của điểm, giá trị từ 0 đến 1.
+ cmap='viridis': Bảng màu dùng để hiển thị màu cho các điểm.
+ NumPy thường được sử dụng cùng với Matplotlib để tạo dữ liệu đầu vào cho biểu đồ.

```python
import numpy as np

# Dữ liệu ngẫu nhiên
x = np.random.rand(50) * 100  # 50 điểm ngẫu nhiên trong khoảng 0 - 100 cho trục X
y = np.random.rand(50) * 100  # 50 điểm ngẫu nhiên trong khoảng 0 - 100 cho trục Y
sizes = np.random.rand(50) * 100  # Kích thước điểm ngẫu nhiên
colors = np.random.rand(50)       # Màu sắc ngẫu nhiên

# Vẽ màu sắc và kích thước khác nhau cho từng điểm
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()  # Hiển thị thanh màu
plt.title("Biểu đồ phân tán nâng cao")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")

plt.show()
```
