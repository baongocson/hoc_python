# Biểu đồ tròn
#### `plt.pie( )`

+ autopct='%1.1f%%': Hiển thị phần trăm cho từng phần
+ startangle=100: Xoay biểu đồ để phần đầu tiên bắt đầu từ góc 100 độ

```python
import matplotlib.pyplot as plt

# dữ liệu cho biểu đồ hình tròn
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['red', 'yellow', 'blue', 'green']


# vẽ biểu đồ hình tròn
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=100)

plt.title("Biểu đồ hình tròn")
plt.show()
```

```python
explode = [0.1, 0, 0, 0]  # Tách phần A ra khỏi biểu đồ

plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode)
plt.title('Biểu đồ hình tròn với phần tách rời')
plt.show()
```

+ `centre_circle = plt.Circle((0,0),0.70,fc='white')`: tạo vong tròn ở giữa

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
# Vẽ vòng tròn ở giữa
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Biểu đồ hình tròn dạng Donut')
plt.show()
```

+ radius: bán kính
+ pctdistance: vị trí các chữ số % so với tâm, mặc định là 0,6

```python
sizes_inner = [10, 20, 30, 40]
sizes_outer = [15, 25, 35, 25]

plt.pie(sizes_outer, radius=1, labels=labels, autopct='%1.1f%%', pctdistance=0.85)
plt.pie(sizes_inner, radius=0.7, autopct='%1.1f%%', pctdistance=0.75)

plt.title('Biểu đồ hình tròn lồng nhau')
plt.show()
```
