# Biểu đồ cột
#### `plt.bar()`: cột dọc
#### `plt.barh()`: cột ngang

```python
import matplotlib.pyplot as plt

# Dữ liệu cho biểu đồ
labels = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Vẽ biểu đồ cột
plt.bar(labels, values, color='skyblue', edgecolor='black')

# Thêm nhãn cho trục x và y
plt.xlabel('Loại')
plt.ylabel('Giá trị')
plt.title('Biểu đồ cột đơn giản')

# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dữ liệu cho biểu đồ
labels = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Vẽ biểu đồ cột nằm ngang
plt.barh(labels, values, color='skyblue', edgecolor='black')

# Thêm nhãn cho trục x và y
plt.xlabel('Giá trị')
plt.ylabel('Loại')
plt.title('Biểu đồ cột nằm ngang')

# Hiển thị biểu đồ
plt.show()
```

+ `plt.xticks(rotation=30)`: xoay 30 trục x
+ edgecolor='black': Thêm viền đen

```python
import matplotlib.pyplot as plt

# các nghề Data 
fields = ['Data Analyst', 'Data Scientist', 'Data Engineer', 'BI Analyst', 'Machine Learning Engineer']
salaries = [15, 25, 28, 20, 30]  # Lương tb

# Vẽ biểu đồ cột
plt.bar(fields, salaries, color='steelblue', edgecolor='black')

# Thêm nhãn cho trục x và y
plt.xlabel('Ngành')
plt.ylabel('Mức lương trung bình (triệu VNĐ/tháng)')
plt.title('Mức lương các ngành Data tại Việt Nam')

# Xoay nhãn trục x để dễ đọc hơn
plt.xticks(rotation=30)

# Hiển thị biểu đồ
plt.show()
```
