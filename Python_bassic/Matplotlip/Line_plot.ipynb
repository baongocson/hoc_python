# Biểu đồ đường:

### `plt.plot(x, y, fmt, **kwargs)`: 
1. x: Danh sách hoặc mảng chứa dữ liệu trên trục x
2. y: Danh sách hoặc mảng chứa dữ liệu trên trục y
3. fmt (tùy chọn): Chuỗi định dạng ngắn gọn cho kiểu dòng, màu sắc, và marker
4. **kwargs: Các tham số khác để tùy chỉnh đồ thị


#### fmt: tham số ngắn gọn cho màu sắc, kiểu marker và kiểu đường 
+ "bo-": Kết hợp màu xanh dương (b), marker dạng tròn (o), và đường liền (-)
+ "r--": Kết hợp màu đỏ (r) với đường đứt (--)

#### **kwargs:
1. linestyle (ls): Kiểu đường nối giữa các điểm. Một số kiểu thường gặp:
+ '-': Đường liền
+ '--': Đường đứt
+ ':': Đường chấm
2. color: Màu của đường
3. linewidth (lw): Độ dày của đường
4. label: Nhãn của đường
...

```python
import matplotlib.pyplot as plt

# dữ liệu cho trục x và y
x = [1, 2, 3, 4]
y = [30, 25, 20, 10]

# vẽ biểu đồ đường
plt.plot(x, y)

# thêm tiêu đề và nhãn trục
plt.title("Biểu đồ đường")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")

# hiển thị biểu đồ
plt.show()
```

+ `plt.grid(True)`: thêm lưới
+ `plt.legend()` : Ghi chú
+ loc = 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
+ bbox_to_anchor=(0, -0.1) vị trí đặt ghi chú


```python
import matplotlib.pyplot as plt

# Tạo dữ liệu tăng dần
x = range(1, 11)  # Trục x từ 1 đến 10
y = [i * 2 for i in x]  # Giá trị y tăng dần gấp đôi mỗi lần

# Vẽ biểu đồ
plt.plot(x, y, marker='o', linestyle='--', color='blue', label='Đường tăng dần')
plt.xlabel('Trục X')
plt.ylabel('Trục Y')
plt.title('Biểu đồ đường luôn tăng')

# Đặt chú thích ở dưới biểu đồ
plt.legend(loc="upper center", bbox_to_anchor=(0, -0.1))
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```
